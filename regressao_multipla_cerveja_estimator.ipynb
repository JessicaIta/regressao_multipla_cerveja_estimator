## Análise de dados com Estimators do Tensorflow

```python
import pandas as pd
base = pd.read_csv('Consumo_cerveja.csv')
base.head()
```

```python
base.columns
```

```python
colunas_usadas = ['Temperatura_Media_C', 'Temperatura_Minima_C',
       'Temperatura_Maxima_C', 'Precipitacao_mm', 'Final_Semana',
       'Consumo_cerveja_litros']
```

```python
colunas_usadas
```

```python
base = pd.read_csv('Consumo_cerveja.csv', usecols = colunas_usadas)
```

```python
base.columns
```

```python
#Normalizando as colunas com dados numéricos
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler()
base[['Temperatura_Media_C', 'Temperatura_Minima_C', 'Temperatura_Maxima_C',
       'Precipitacao_mm']] = scaler_x.fit_transform(base[['Temperatura_Media_C', 'Temperatura_Minima_C', 
                                                          'Temperatura_Maxima_C','Precipitacao_mm']])
```

```python
base.head()
```

```python
#Normalizando a coluna previsora
scaler_y = MinMaxScaler()
base[['Consumo_cerveja_litros']] = scaler_y.fit_transform(base[['Consumo_cerveja_litros']])
```

```python
base.head()
```

```python
#Atribuindo os valor de X (inputs) e y(previsão)
X = base.drop('Consumo_cerveja_litros', axis = 1)
y = base.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
type(X)
```

```python
type(y)
```

```python
#Colunas para o input
input_colunas = colunas_usadas[0:5]
input_colunas
```

```python
#Importando o tensorflow
import tensorflow as tf
```

```python
colunas = [tf.feature_column.numeric_column(key = c) for c in input_colunas]
```

```python
print(colunas[2])
```

```python
#Separando em conjunto de treino e teste(30%) dos dados
from sklearn.model_selection import train_test_split
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size = 0.3)
```

```python
X_treinamento.shape
```

```python
y_treinamento.shape
```

```python
X_teste.shape
```

```python
y_teste.shape
```

```python
#função de treinamento (input pandas) com lote de 32 registros e input aleatório dos dados
funcao_treinamento = tf.estimator.inputs.pandas_input_fn(x = X_treinamento, y = y_treinamento,
                                                        batch_size = 32, num_epochs = None, shuffle = True)
```

```python
#função de teste  (input pandas) com lote de 32 registros e input em ordem dos dados
funcao_teste = tf.estimator.inputs.pandas_input_fn(x = X_teste, y = y_teste,
                                                   batch_size = 32, num_epochs = 1000, shuffle = False)
```

```python
#Criando o regressor linear com as colunas escolhidas
regressor = tf.estimator.LinearRegressor(feature_columns=colunas)
```

```python
#Treinamento do regressor com 1000 steps
regressor.train(input_fn=funcao_treinamento, steps = 1000)
```

```python
#avaliação do treinamento
metricas_treinamento = regressor.evaluate(input_fn=funcao_treinamento, steps = 1000)
```

```python
#Avaliação do teste
metricas_teste = regressor.evaluate(input_fn=funcao_teste, steps = 1000)
```

```python
metricas_treinamento
```

```python
metricas_teste
```

```python
#função de previsão utilizando os dados do teste
funcao_previsao = tf.estimator.inputs.pandas_input_fn(x = X_teste, shuffle = False)
```

```python
previsoes = regressor.predict(input_fn=funcao_previsao)
```

```python
list(previsoes)
```

```python
#Listando os valores das previsões
valores_previsoes = []
for p in regressor.predict(input_fn=funcao_previsao):
    valores_previsoes.append(p['predictions'])
```

```python
valores_previsoes
```

```python
#Realizando reshape para transformar na escala inicial
import numpy as np
valores_previsoes = np.asarray(valores_previsoes).reshape(-1,1)
```

```python
valores_previsoes
```

```python
#Retornando a escala inicial
valores_previsoes = scaler_y.inverse_transform(valores_previsoes)
valores_previsoes
```

```python
y_teste.shape
```

```python
#Comparar as previsões realizadas com os dados de teste
y_teste2 = y_teste.values.reshape(-1,1)
y_teste2.shape
```

```python
y_teste2 = scaler_y.inverse_transform(y_teste2)
```

```python
y_teste2
```

```python
#Calculando o Erro Absoluto Médio e o Erro Quadrático Médio 
from sklearn.metrics import mean_absolute_error,mean_squared_error
mae = mean_absolute_error(y_teste2, valores_previsoes)
mse = mean_squared_error(y_teste2, valores_previsoes)
```
